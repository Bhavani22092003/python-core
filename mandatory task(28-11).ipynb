## Exception Handling

* Exception-->Error,rectifying an error  
* It works only on <b>syntax errors</b> ,not on <b>logical errors.</b>



<b>Try</b>   --->> checking the logic or try block

<b>Except</b>  ---->> if try block is false then it prints except block.we can also give in paranthesis() ,that what type of error can be occured in try block  

<b>Else</b> ----->> You can use the else clause to specify an action to be executed if no exception occurs. If an exception does occur and is caught by an except clause, the action in the else clause will not be executed.

<b>Find</b> It defines a block of code to run when the try... except...else block is final. The finally block will be executed no matter if the try block raises an error or not.

```python
import sys,traceback
from sys import exc_info
```

##### Trace back the errors :
Trace back--->it trace back the exact location where the error was occured

```python
try:
    x,y=int(input('enter x: ')),int(input('enter x: '))
    z=x/y
    print(z)
    
except:
    print(f'{exc_info()[0]} error occured....')
    print(traceback.print_exc())
    print('pass  valid inputs')
```

### dynamic code for try and except 

```python
while True:
    try:
        x,y=int(input('enter x: ')),int(input('enter x: '))
        z=x/y
        print(z)
        break
    except:
        print(f'{exc_info()[0]} error occured....')
        print(traceback.print_exc())
        print('pass  valid inputs')
```

#### Another way for the above program

```python
def foo():
        try:
            x,y=int(input('enter x: ')),int(input('enter x: '))
            z=x/y
            print(z)
        except:
            print(f'{exc_info()[0]} error occured....')
            print(traceback.print_exc())
            print('pass  valid inputs')
    
```

```python
class LevelOneError(Exception):
    def __init__(self):
        print('trying to access director files ')
class LevelTwoError(Exception):
    def __init__(self):
        print('trying to access director and manage files')
class CredError(Exception):
    def __init__(self):
        print('trying to access all files')
```

```python
empClearence = int(input('enter your clearence level: '))
fileName =input('enter the file name you are tryong to access with it\s extension: ')
if (empClearence in [3,2,1]) and  ( fileName.split('.')[1] in ['emp', 'mang', 'dir']):
    if (empClearence == 3) and (fileName != 'emp'):
        raise LevelTwoError()
    elif (empClearence == 2) and (fileName == 'dir'):
        raise LevelOneError()
    else:
        print('Access Granted')
else:
    raise CredError()
```
